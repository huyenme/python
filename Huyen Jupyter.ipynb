```python
import altair as alt
```

```python
alt.renderers.enable('notebook')
```

```python
employers = my_contribs.groupby("contributor_employer").amount.sum(). reset_index().sort_values("amount", ascending=False).head()
```

```python
employers
```

```python
alt.Chart(employers).mark_bar().encode(
    x="amount",
    y="contributor_employer",
    )
```

```python
alt.Chart(employers, title="Top Contributors").mark_bar().encode(
    x=alt.X("amount", title="Amount of Contribution"),
    y=alt.Y(
        'contributor_employer', title="Contributor Employers",
        sort=alt.EncodingSortField(
            field="amount",  # The field to use for the sort
            op="sum",  # The operation to run on the field prior to sorting
            order="descending"  # The order to sort in
 )))
```
