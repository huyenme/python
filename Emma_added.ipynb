```python
top_supporters["contributor_fullname"] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
barh = top_supporters.amount.plot.barh(color="red", legend=True, alpha=.5)
barh.set_yticklabels(top_supporters.contributor_fullname)
```

```python
top_supporters.amount.plot.pie()
```

```python
def combine_names(row):
    if row.contributor_fullname.startswith("SEAN PARKER"):
        return "SEAN PARKER"
    return row.contributor_fullname
```

```python
top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters["contributor_cleanname"] = top_supporters.apply(combine_names, axis=1)
```

```python
top_supporters.head()
```

```python
top_new= top_supporters.groupby(
    "contributor_cleanname"
).amount.sum().reset_index().sort_values("amount", ascending=False)
```

```python
top_new
```

```python
chart=top_new.amount.plot.barh()
chart.set_yticklabels(top_new.contributor_cleanname).color(Red)
```

```python
import altair as alt
```

```python
alt.renderers.enable("notebook")
```

```python
bars= alt.Chart(top_new, title="Top Contributors").mark_bar(color="red").encode(
    x=alt.X("amount", title="Amount of Contribution", ),
    y=alt.Y(
        'contributor_cleanname', title="Contributor Names",
        sort=alt.EncodingSortField(
            field="amount",  # The field to use for the sort
            op="sum",  # The operation to run on the field prior to sorting
            order="descending"  # The order to sort in
 )))
text = bars.mark_text(
    align='left',
    baseline='middle',
    dx=3  # Nudges text to right so it doesn't appear on top of the bar
).encode(
    text='amount:Q'
)

(bars + text).properties(height=100)
```
